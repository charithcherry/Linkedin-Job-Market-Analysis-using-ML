```python
from bs4 import BeautifulSoup

# Assuming 'html_content' contains the HTML data you pasted above.
html_content = response.text

# Parse the HTML
soup = BeautifulSoup(html_content, 'html.parser')

# Find all job postings
job_cards = soup.find_all('div', class_='base-card')

jobs = []
for job in job_cards:
    # Extract title
    title = job.find('h3', class_='base-search-card__title')
    title_text = title.get_text(strip=True) if title else 'N/A'
    
    # Extract company name
    company = job.find('h4', class_='base-search-card__subtitle')
    company_text = company.get_text(strip=True) if company else 'N/A'
    
    # Extract job location
    location = job.find('span', class_='job-search-card__location')
    location_text = location.get_text(strip=True) if location else 'N/A'
    
    # Extract job posting date
    post_date = job.find('time')
    post_date_text = post_date.get_text(strip=True) if post_date else 'N/A'
    
    # Extract job URL
    job_link = job.find('a', class_='base-card__full-link')
    job_url = job_link['href'] if job_link else 'N/A'

    jobs.append({
        'Title': title_text,
        'Company': company_text,
        'Location': location_text,
        'Posted': post_date_text,
        'URL': job_url
    })

# Display jobs data
for job in jobs:
    print(f"Title: {job['Title']}")
    print(f"Company: {job['Company']}")
    print(f"Location: {job['Location']}")
    print(f"Posted: {job['Posted']}")
    print(f"URL: {job['URL']}\n")
```

```python
def generate_url(keywords=None, location=None, start=0):
    base_url = "https://yourserver.com/search?start={}"

    # Start with the base URL and add the start parameter
    formatted_url = base_url.format(start)
    
    # Check if keywords exist and append to the URL
    if keywords:
        formatted_url += "&keywords={}".format(keywords.replace(" ", "%20"))

    # Check if location exists and append to the URL
    if location:
        formatted_url += "&location={}".format(location.replace(" ", "%20"))

    return formatted_url
```

```python
import requests
from bs4 import BeautifulSoup
import math
import pandas as pd

def scrape_linkedin_jobs(base_url, num_jobs=117):
    l = []  # List to store job IDs
    k = []  # List to store job details
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36"
    }

    # Loop to scrape job IDs from the main URL
    for i in range(0, math.ceil(num_jobs / 25)):
        target_url = base_url.format(i)  # Format URL with pagination
        res = requests.get(target_url, headers=headers)
        
        # Check for successful response
        if res.status_code != 200:
            print(f"Failed to retrieve data from {target_url}")
            continue
        
        soup = BeautifulSoup(res.text, 'html.parser')
        alljobs_on_this_page = soup.find_all("li")
        print(f"Number of jobs on page {i}: {len(alljobs_on_this_page)}")
        
        for job in alljobs_on_this_page:
            jobid = job.find("div", {"class": "base-card"}).get('data-entity-urn').split(":")[3]
            l.append(jobid)  # Append job ID to list

    # Loop to scrape job details using job IDs
    for job_id in l:
        target_url = f'https://www.linkedin.com/jobs-guest/jobs/api/jobPosting/{job_id}'
        resp = requests.get(target_url, headers=headers)
        
        # Check for successful response
        if resp.status_code != 200:
            print(f"Failed to retrieve data for job ID: {job_id}")
            continue
            
        soup = BeautifulSoup(resp.text, 'html.parser')
        o = {}  # Dictionary to store job details

        # Scraping job details with error handling
        try:
            o["company"] = soup.find("div", {"class": "top-card-layout__card"}).find("a").find("img").get('alt')
        except:
            o["company"] = None

        try:
            o["job-title"] = soup.find("div", {"class": "top-card-layout__entity-info"}).find("a").text.strip()
        except:
            o["job-title"] = None

        try:
            o["level"] = soup.find("ul", {"class": "description__job-criteria-list"}).find("li").text.replace("Seniority level", "").strip()
        except:
            o["level"] = None

        k.append(o)  # Append job details to the list

    # Create DataFrame and save to CSV
    df = pd.DataFrame(k)

    return df
```
